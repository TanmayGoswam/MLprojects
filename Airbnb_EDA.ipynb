**Project summary**

Since its inception in 2008, Airbnb has revolutionized the travel industry by offering unique and personalized lodging experiences. The platform connects guests with hosts who offer a variety of accommodations, from spare rooms to entire homes.
The main goal is to do data analysis on the Airbnb dataset, to do some
visualizations projecting the nature of the dataset, and also to provide key insights into various aspects of Airbnb listings in New York City.
The key columns in the dataset are:

     id: Unique identifier for each listing
     name: Name of the listing
     host_id: Unique identifier for each host
     host_name: Name of the host
     neighbourhood_group: Geographic region of the listing (e.g., Manhattan, Brooklyn)
     neighbourhood: Specific neighborhood within the region
     latitude and longitude: Geographical coordinates of the listing
     room_type: Type of room offered (e.g., Entire home/apt, Private room, Shared room)
     price: Price per night in USD
     minimum_nights: Minimum number of nights a guest can stay
     number_of_reviews: Total number of reviews the listing has received
     last_review: Date of the last review
     reviews_per_month: Average number of reviews per month
     calculated_host_listings_count: Number of listings a host has
     availability_365: Number of days the listing is available in a year.

 ### 2.Data Cleaning


Checking for Missing Values

```python
# Check for missing values
missing_values = df.isnull().sum()
missing_values
```

**Handling Missing Values**

```python
# Fill missing values in 'reviews_per_month' with 0
df['reviews_per_month'].fillna(0, inplace=True)
```

```python
# Fill missing values in 'last_review' with a placeholder date
df['last_review'].fillna('1900-01-01', inplace=True)
```

**Checking for Duplicates and remove**

```python
# Check for duplicates
duplicates = df.duplicated().sum()
duplicates
```

```python
# Remove duplicates if any
df.drop_duplicates(inplace=True)
```

### Bivariate analysis

Numerical-numerical

```python
#scatterplot
plt.scatter(df['number_of_reviews'],df['price'])
plt.xlabel('number_of_reviews')
plt.ylabel('price')
plt.show()
```

Numerical-categorical

```python
sns.barplot(x='room_type',y='price',data=df)
```

Categorical-categorical

```python
#to see data in table format with respect to percentage
pd.crosstab(df['room_type'],df['neighbourhood_group'],normalize='columns')*100
```

```python
#graph heatmap

sns.heatmap(pd.crosstab(df['room_type'],df['neighbourhood_group'],normalize='columns')*100,annot=True,cmap='coolwarm')
```

### Multivariate analysis
